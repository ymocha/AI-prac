```python
#1-1
#폐암환자 생존 예측 Thoraci

# 딥러닝을 구성하는 데 필요한 케라스 함수 호출
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# 필요한 라이브러리 불러오기
import numpy as np
import tensorflow as tf

# 실행할 때마다 같은 결과를 출력 
np.random.seed(3)
tf.random.set_seed(3)

# 준비된 수술 환자 데이터를 불러오기
Data_set = np.loadtxt("/Users/yk/Desktop/AIprac/ThoraricSurgery.csv", delimiter=",") # PC내부의 검색기능을 활용하여 파일을 찾은뒤 옵션키로 위치를 복사해서 붙여넣는다.

# 환자의 기록과 수술결과를 X, Y로 구분하여 저장
X = Data_set[:, :17] # 전체차원에서 0부터 16번째 인덱스의 데이터를 X로 지정(feature)
Y = Data_set[:, 17] # 전체차원에서 17째 인덱스의 데이터를 Y로 지정(label)

# 딥러닝 구조를 결정(모델을 설정하고 실행) 
# 두개의 층을 쌓음
model = Sequential()
model.add(Dense(30, input_dim=17, activation="relu"))
model.add(Dense(1, activation="sigmoid"))
#activation: 활성함수. 다음 층으로 어떻게 값을 넘길지 결정하는 부분

# 딥러닝 실행
model.compile(loss="binary_crossentropy", optimizer ="adam", metrics=["accuracy"]) 
model.fit(X, Y, epochs=100, batch_size=10)
#loss: 한 번 신경망이 실행될 때마다 오차 값을 추적하는 함수
#optimizer: 오차를 어떻게 줄여 나갈지 정하는 함수
```
